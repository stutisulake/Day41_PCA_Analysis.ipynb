
# Work Summary
• Applied Principal Component Analysis (PCA) for dimensionality reduction.
• Identified redundancy and noise in high-dimensional data.
• Standardized features to ensure equal contribution across variables.
• Used Scree Plot to determine optimal number of components.
• Reduced dataset from 13 features to 3 principal components retaining ~96% variance.
• Visualized transformed data using 2D scatter plot to reveal clusters and patterns.

# Learnings / Outcomes
• Variance = Information: PCA searches for directions in which data "stretches" the most because that is where the most unique information is found.
• New Features, Not Old Ones: PCA combines columns into completely new, uncorrelated variables rather than selecting existing ones.
• Orthogonality: Every Principal Component is mathematically independent (90-degree angle), which helps remove multicollinearity.

# Blockers / Risks
• Interpretability Trade-off: PCA can make data harder to understand; "PC1" is a combination of features, not a straightforward variable.
• Loss of Detail: Reducing too many dimensions can cause clusters to overlap, requiring careful selection of components using Scree Plot.

# Skills
• Principal Component Analysis (PCA) – Technique to transform high-dimensional correlated features into a smaller set of uncorrelated components.
• Dimensionality Reduction – Reducing the number of features while retaining most of the information to simplify analysis and modeling.
• Data Standardization (Feature Scaling) – Adjusting features to the same scale so no single feature dominates the variance.
• Explained Variance Analysis – Measuring how much information each principal component captures from the original data.
• Scree Plot Interpretation – Visual tool to decide the optimal number of principal components by showing variance explained by each component.
• Data Visualization (2D Scatter Plot) – Plotting high-dimensional data in 2D to reveal clusters and patterns.
• Multicollinearity Handling – Removing correlations among features to prevent redundancy in models.
• Linear Algebra Intuition (Orthogonality, Variance) – Understanding mathematical relationships between components and directions of maximum variance.
• Feature Transformation – Creating new variables (components) that summarize the original dataset effectively.
• Analytical Thinking – Applying logic and reasoning to interpret PCA results and make informed decisions.

# Conclusion
• PCA is effective for reducing high-dimensional data while preserving most important information.
• Standardization ensures fair contribution from all features and avoids bias.
• Scree Plot helps identify the optimal number of components for a good balance between reduction and information retention.
• Dimensionality reduction simplifies visualization and reveals hidden patterns like clusters.
• There is a trade-off between reduced dimensionality and interpretability that requires careful analysis.

# Trade-off: Reduced Dimensionality vs. Interpretability
**Reduced Dimensionality Benefits**
• Simplifies datasets by decreasing the number of features.
• Retains most of the original information through principal components.
• Helps remove redundancy and multicollinearity.
• Enables easier visualization of clusters and patterns (e.g., 2D scatter plots).
• Can improve performance of machine learning models by reducing noise.

**Interpretability Challenges**
• Principal components are combinations of original features, making them harder to interpret.
• The meaning of "PC1", "PC2", etc., is abstract rather than concrete.
• Excessive dimensionality reduction can obscure subtle patterns or details.
• Decisions based solely on PCA-transformed features may be less intuitive.
• Requires careful analysis using tools like Scree Plot to balance variance retention and simplicity.

**Key Insight**
• There is a **trade-off**: more reduction simplifies the dataset but reduces interpretability.
• Optimal component selection is essential to maintain sufficient information while keeping the transformed data understandable.
